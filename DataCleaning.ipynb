```python
# ============================================
# Step 9B: Remove outliers
# ============================================

print(f"Rows before outlier removal: {len(df)}")

# --- Price Outliers ---
# Remove properties with price = 0 or negative
df = df[df['Price'] > 0]

# Remove top 1% extreme prices (likely errors)
price_99 = df['Price'].quantile(0.99)
price_01 = df['Price'].quantile(0.01)
print(f"Price range keeping: ₹{price_01:,.0f} to ₹{price_99:,.0f}")
df = df[(df['Price'] >= price_01) & (df['Price'] <= price_99)]

# --- Area Outliers ---
# Minimum realistic flat size = 100 sqft
# Maximum realistic flat size = 15000 sqft
df = df[(df['Area'] >= 100) & (df['Area'] <= 15000)]

# --- Bedroom Outliers ---
# Keep 1 to 10 bedrooms only
df = df[(df['No. of Bedrooms'] >= 1) &
        (df['No. of Bedrooms'] <= 10)]

print(f"Rows after outlier removal: {len(df)}")
print(f"Rows removed: {original_count - len(df)}")
```

```python
# ============================================
# Step 10: Clean text columns
# ============================================

# 1. Strip whitespace from all text columns
df['Location'] = df['Location'].str.strip()
df['City'] = df['City'].str.strip()

# 2. Standardize city names (fix capitalization)
df['City'] = df['City'].str.title()
# "mumbai" → "Mumbai", "BANGALORE" → "Bangalore"

# 3. Check unique cities
print("CITIES IN DATASET:")
print(df['City'].value_counts())
print()

# 4. Fix any spelling variations you find
# (adjust these based on what you actually see)
city_fixes = {
    'Bengaluru': 'Bangalore',
    'bengaluru': 'Bangalore',
    'bangalore': 'Bangalore',
    'mumbai': 'Mumbai',
    'kolkata': 'Kolkata',
    'delhi': 'Delhi'
}
df['City'] = df['City'].replace(city_fixes)

print("\nCITIES AFTER FIXING:")
print(df['City'].value_counts())
```

```python
# ============================================
# Step 11: Create new useful columns
# ============================================

# 1. Price per sqft (MOST IMPORTANT)
df['price_per_sqft'] = df['Price'] / df['Area']
print("✅ price_per_sqft created")

# 2. Area per bedroom
df['area_per_bedroom'] = df['Area'] / df['No. of Bedrooms']
print("✅ area_per_bedroom created")

# 3. Luxury Score
luxury_cols = ['Gymnasium', 'SwimmingPool', 'GolfCourse',
               'ClubHouse', 'LandscapedGarden']
# Only use columns that exist in your dataset
luxury_cols = [c for c in luxury_cols if c in df.columns]
df['luxury_score'] = df[luxury_cols].sum(axis=1)
print(f"✅ luxury_score created (from {len(luxury_cols)} features)")

# 4. Convenience Score
conv_cols = ['ATM', 'ShoppingMall', 'Hospital',
             'School', 'CarParking', 'LiftAvailable']
conv_cols = [c for c in conv_cols if c in df.columns]
df['convenience_score'] = df[conv_cols].sum(axis=1)
print("✅ convenience_score created")

# 5. Furnished Score (appliances)
furnish_cols = ['WashingMachine', 'AC', 'Wifi', 'TV',
                'Microwave', 'Refrigerator', 'DiningTable',
                'Sofa', 'Wardrobe']
furnish_cols = [c for c in furnish_cols if c in df.columns]
df['furnished_score'] = df[furnish_cols].sum(axis=1)
print("✅ furnished_score created")

# 6. Total amenity score
all_score_cols = luxury_cols + conv_cols + furnish_cols
df['total_amenity_score'] = df[all_score_cols].sum(axis=1)
print("✅ total_amenity_score created")

# 7. Property type label
df['property_type'] = df['Resale'].map({0: 'New', 1: 'Resale'})
print("✅ property_type created")
```
